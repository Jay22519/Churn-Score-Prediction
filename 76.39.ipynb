```python
def CountFrequency(lista):
    my_list = lista.copy()        
    for i in range(len(my_list)) :
        if(str(my_list[i]) == 'nan') :
            my_list[i] = 'Unk'
            
    #print(my_list)
    my_list.sort()
    
    # Creating an empty dictionary 
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
  
    for key, value in freq.items():
        print (key," : ",value)
```

```python
def clean_data(path) :
    
    #Done with input
    data = pd.read_csv(path) 
    
    try :
        data = data[data.churn_risk_score != -1]
    except :
        print("This is a test data")
        
    data = data.reset_index()
    ### First enchancing age feature 
    
    data['Age_type'] = [0 for i in range(len(data))]
    
    for i in range(data.shape[0]) :
        if(data.age[i] >=10 and data.age[i] <=20) :
            data['Age_type'][i] = 1 
        elif(data.age[i] >=21 and data.age[i] <=31) :
            data['Age_type'][i] = 2 
        elif(data.age[i] >= 32 and data.age[i] <=42) :
            data['Age_type'][i] = 3 
        
        elif(data.age[i] >= 43 and data.age[i] <=53) :
            data['Age_type'][i] = 4 
        else :
            data['Age_type'][i] = 5 
            
            
    print("Done with age .......\n")   
    #### Enhancing region_category 
    data['region_type'] = [0 for i in range(len(data))]
    for i in range(data.shape[0]) :
        
        if(str(data.region_category[i]) == 'nan') :
            data.region_category[i] = 0
        else :
            data['region_type'][i] = 1 
            if(data.region_category[i] == 'City') :
                data.region_category[i] = 1 
            elif(data.region_category[i] == 'Town') :
                data.region_category[i] = 2 
            elif(data.region_category[i] == 'Village') :
                data.region_category[i] = 3 
                
    print("Done with region ........\n")      
    
    
    ## Encoding gender 
    
    for i in range(data.shape[0]) :
        if(data.gender[i] == 'F') :
            data.gender[i] = 0 
        elif(data.gender[i] == 'M') :
            data.gender[i] = 1 
        else :
            data.gender[i] = 2 
            
    
    #data = data.drop(['security_no'],axis = 1)
    
    data['secur_type'] = [0 for i in range(len(data))] 
    for i in range(len(data)) :
        data['secur_type'][i] = data.security_no[i][:1]


    di = {}
    counter =  0 
    for i in range(len(data)) :
        try :
            a = di[data.secur_type[i]]
        except :
            di[data.secur_type[i]] = counter 
            counter += 1

    for i in range(len(data)) :
        data['secur_type'][i] = int(di[data['secur_type'][i]])
    
    print("Done with security and Gender........\n")
    
    
    
    ### Enhancing Membership type 
    data['mem_15'] = [0 for i in range(len(data))] 
    data['mem_23'] = [0 for i in range(len(data))]
    data['mem_4'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        if(data.membership_category[i] == 'Basic Membership') :
            data['mem_15'][i] = 1
            data['mem_23'][i] = 1 
            data['mem_4'][i] = 0 
            data.membership_category[i] = 0
            
            
        elif(data.membership_category[i] == 'Gold Membership') :
            data['mem_15'][i] = 0
            data['mem_23'][i] = 0 
            data['mem_4'][i] = 0 
            data.membership_category[i] = 3
            
            
        elif(data.membership_category[i] == 'No Membership') :
            data['mem_15'][i] = 1
            data['mem_23'][i] = 1
            data['mem_4'][i] = 0 
            data.membership_category[i] = 1
            
            
        elif(data.membership_category[i] == 'Platinum Membership') :
            data['mem_15'][i] = 0
            data['mem_23'][i] = 0
            data['mem_4'][i] = 1 
            data.membership_category[i] = 4
            
            
        elif(data.membership_category[i] == 'Premium Membership') :
            data['mem_15'][i] = 0
            data['mem_23'][i] = 0 
            data['mem_4'][i] = 1
            data.membership_category[i] = 5
            
            
        elif(data.membership_category[i] == 'Silver Membership') :
            data['mem_15'][i] = 1
            data['mem_23'][i] = 0 
            data['mem_4'][i] = 0 
            data.membership_category[i] =  2
            
            
        else :
            data['mem_15'][i] = 2
            data['mem_23'][i] = 2
            data['mem_4'][i] = 2
            data.membership_category[i] = 6 
            
            
    print("Done with membership ..........\n") 
            
    ## Enhancing join date and breaking it into year and month 
    data['month'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        data['month'][i] = data['joining_date'][i][5:7]
        data['joining_date'][i] = data['joining_date'][i][0:4]
        
    print("Done with month and year .............\n")
    
    
    
    ## Enhancing referral_joined
    data['ref_type'] = [0 for i in range(len(data))]
    for i in range(len(data)) :
        if(data['joined_through_referral'][i]== 'No') :
            data['joined_through_referral'][i] = 0 
            data['ref_type'][i] = 1
        elif(data['joined_through_referral'][i]== 'Yes') :
            data['joined_through_referral'][i] = 1
            data['ref_type'][i] = 1
        else :
            data['joined_through_referral'][i] = 2 
            data['ref_type'][i] = 0 
            
    print("Done with referral ................\n")
            
        
    ### Enhancing day of last login 
    
    data['days_login_type'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        
        if(data['days_since_last_login'][i] == -999) :
            data['days_login_type'][i] = 0 
        elif(data['days_since_last_login'][i] == 1 or data['days_since_last_login'][i] == 2) :
            data['days_login_type'][i] = 1 
        elif(data['days_since_last_login'][i] > 2 and data['days_since_last_login'][i] <=24) :
            data['days_login_type'][i] = 2 
        else :
            data['days_login_type'][i] = 3 
            
    
    print("Done with days of last login .............\n")
    
    #### Encoding preferred_offer_types
    
    for i in range(len(data)) :
        if(data['preferred_offer_types'][i] == 'Credit/Debit Card Offers') :
            data['preferred_offer_types'][i] = 0 
        elif(data['preferred_offer_types'][i] == 'Gift Vouchers/Coupons') :
            data['preferred_offer_types'][i] = 1
        elif(data['preferred_offer_types'][i] == 'Without Offers') :
            
            data['preferred_offer_types'][i] = 2 
        else :
            data['preferred_offer_types'][i] = 3 
            
    print("Done with preferred offer types ..............\n")
    
    
    #### Enhancing medium of operation and creating med_both feature which is 1 if customer uses both mobile and laptop 
    
    data['med_both'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        if(data['medium_of_operation'][i] == 'Desktop') :
            data['medium_of_operation'][i] = 0 
        elif(data['medium_of_operation'][i] == 'Smartphone') :
            data['medium_of_operation'][i] = 1 
        elif(data['medium_of_operation'][i] == 'Both') :
            data['medium_of_operation'][i] = 2 
            data['med_both'][i] = 1
        else :
            data['medium_of_operation'][i] = 3 
            
            
    print("Done with medium of operations .............\n") 
    
    
    ### Encoding Internet options 
    
    for i in range(len(data)) :
        if(data['internet_option'][i] == 'Fiber_Optic') :
            data['internet_option'][i] = 0 
        elif(data['internet_option'][i] == 'Mobile_Data') :
            data['internet_option'][i] = 1
        else :
            data['internet_option'][i] = 2 
            
    print("Done with internet options ..................\n")
        
        
    ### Changing last time visit to hours
    for i in range(len(data)) :
        
        data['last_visit_time'][i] = int(data['last_visit_time'][i][:2])
        
    print("Done with last time visit ..............\n")
    
    
    
    ### Average time spent 
    
    data['avg_time_spent_type'] = [0 for i in range(len(data))]
    for i in range(len(data)) :
        if(data['avg_time_spent'][i]<0) :
            data['avg_time_spent_type'][i] = -1 
        elif(data['avg_time_spent'][i] <100 and data['avg_time_spent'][i]>0) :
            data['avg_time_spent_type'][i] = 1 
        elif(data['avg_time_spent'][i]>=100 and data['avg_time_spent'][i]<200) :
            data['avg_time_spent_type'][i] = 2
        else :
            data['avg_time_spent_type'][i] = 3
            
            
    print("Done with avg_time_spent................\n")
    
    ## Enhancing avg_transaction_val 
    data['avg_transaction_value_type'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        if(data['avg_transaction_value'][i]<25000) :
            data['avg_transaction_value_type'][i] = 0
        elif (data['avg_transaction_value'][i]>25000 and data['avg_transaction_value'][i] < 50000) :
            data['avg_transaction_value_type'][i] = 1 
        else :
            data['avg_transaction_value_type'][i] = 2 
            
    
    print("Done with avg tranction value ..............\n")
    
    
    ### Changing avg_frequency_login_day 
    for i in range(len(data)) :
        try :
            data['avg_frequency_login_days'][i] = float(data['avg_frequency_login_days'][i])
        except :
             data['avg_frequency_login_days'][i] = -100

                
    data['freq_12'] = [0 for i in range(len(data))]
    data['freq_345'] = [0 for i in range(len(data))]
    
    for i in range(len(data)) :
        if(data['avg_frequency_login_days'][i] >= 1 and data['avg_frequency_login_days'][i] <=19) :
            data['freq_12'][i] = 1 
        if(data['avg_frequency_login_days'][i] >=5 and data['avg_frequency_login_days'][i] <=29) :
            data['freq_345'][i] = 1 
            
    print("Done with avg_freq_login_day....................\n")
    
    
    ### Enhancing points_in_wallet 
    data['wallet_1'] = [0 for i in range(len(data))] 
    data['wallet_2'] = [0 for i in range(len(data))] 
    data['wallet_3'] = [0 for i in range(len(data))] 
    data['wallet_4'] = [0 for i in range(len(data))] 
    data['wallet_5'] = [0 for i in range(len(data))] 
    
    points = data.points_in_wallet.copy()
    
    for i in range(len(data)) :
        if(str(points[i]) == 'nan') :
            points[i] = -999 
        else :
            points[i] = float(points[i])

    points = [i//5 for i in points]
    
    for i in range(len(points)) :
        if(points[i] >= 150 and points[i] <=169) :
            data['wallet_1'][i] = 1 
        if(points[i] >= 140 and points[i] <=159) :
            data['wallet_2'][i] = 1 
            data['wallet_3'][i] = 1 
        
        if(points[i] >=120 and points[i] <=143) :
            data['wallet_4'][i] = 1 
        if(points[i] >= 100 and points[i] <=139) :
            data['wallet_5'][i] = 1 
    
    
    data = data.reset_index()
    for i in range(len(data)) :
        if(str(data['points_in_wallet'][i]) == 'nan') :
            data['points_in_wallet'][i] = -999 
        else :
            data['points_in_wallet'][i] = float(data['points_in_wallet'][i])

    data['points_in_wallet'] = [i//100 for i in data['points_in_wallet']]
    
     #data['wallet_1'] = [0 for i in range(len(data))] 
        
    
        
    for i in range(len(data)) :
        if(data['points_in_wallet'][i] < 0) :
            data['points_in_wallet'][i] = 0 
        elif(data['points_in_wallet'][i] > 0 and data['points_in_wallet'][i] <7) :
            data['points_in_wallet'][i] = 1 
        elif(data['points_in_wallet'][i] == 7) :
            data['points_in_wallet'][i] = 2 
        elif(data['points_in_wallet'][i] == 8) :
            data['points_in_wallet'][i] = 3 
        else :
            data['points_in_wallet'][i] = 4
            
            
    print("Done with points in wallet")
    
    
    ### Changin_used_special_discount 
    
    for i in range(len(data)) :
        if(data['used_special_discount'][i] == 'No') :
            data['used_special_discount'][i] = 0 
        elif(data['used_special_discount'][i] == 'Yes') :
            data['used_special_discount'][i] = 1 
        else :
            data['used_special_discount'][i] = 2 
            
    print("Done with used_special_discount ............\n") 
    
    
    ## Encoding offer application preference 
    for i in range(len(data)) :
        if(data.offer_application_preference[i] == 'No') :
            data.offer_application_preference[i] = 0 
        elif(data.offer_application_preference[i] == 'Yes') :
            data.offer_application_preference[i] = 1 
        else :
            data.offer_application_preference[i] = 2 
            
    print("Done with offer application preference............\n")
    
    ## Past complaint 
    for i in range(len(data)) :
        if(data['past_complaint'][i] == 'No') :
            data['past_complaint'][i] = 0 
        elif(data['past_complaint'][i] == 'Yes') :
            data['past_complaint'][i] = 1 
        else :
            data['past_complaint'][i] = 2 
            
    print("Done with past_complaint ...................\n")
    
    
    ### Encoding Complaing status 
    
    for i in range(len(data)) :
        if(data['complaint_status'][i] == 'No Information Available') :
            data['complaint_status'][i] = 0 
        elif(data['complaint_status'][i] == 'Not Applicable') :
            data['complaint_status'][i] = 1 
        elif(data['complaint_status'][i] == 'Solved') :
            data['complaint_status'][i] = 2 
        elif(data['complaint_status'][i] == 'Solved in Follow-up') :
            data['complaint_status'][i] = 3 
        elif(data['complaint_status'][i] == 'Unsolved') :
            data['complaint_status'][i] =  4 
            
            
    print("Done with complaint status ...................\n") 
    
    
    #### Finally encoding and enhancing Feeback :) 
    ##### Feedback_type =0 for Products always in Stock , Quality Customer Care , Reasonable Price ,User Friendly Website 
    ### Feedback_type = 1 for No reason specified  :  2078
           ### Poor Customer Service  :  2064
           ### Poor Product Quality  :  2025
           ### Poor Website  :  2098
            ###Too many ads
    
    data['feedback_type'] = [0 for i in range(len(data))]
    for i in range(len(data)) :
        if(data['feedback'][i] == 'Products always in Stock') :
            data['feedback'][i] = 0 
            data['feedback_type'][i] = 0 
            
        elif(data['feedback'][i] == 'Quality Customer Care') :
            data['feedback'][i] = 1 
            data['feedback_type'][i] = 0 
        elif(data['feedback'][i] == 'Reasonable Price') :
            data['feedback'][i] = 2 
            data['feedback_type'][i] = 0 
        elif(data['feedback'][i] == 'User Friendly Website') :
            data['feedback'][i] = 3 
            data['feedback_type'][i] = 0 
            
        elif(data['feedback'][i] == 'No reason specified') :
            data['feedback'][i] = 4
            data['feedback_type'][i] = 1 
            
        elif(data['feedback'][i] == 'Poor Customer Service') :
            data['feedback'][i] = 5 
            data['feedback_type'][i] = 1             
        elif(data['feedback'][i] == 'Poor Product Quality') :
            data['feedback'][i] = 6 
            data['feedback_type'][i] = 1             
        elif(data['feedback'][i] == 'Poor Website') :
            data['feedback'][i] = 7 
            data['feedback_type'][i] = 1             
        elif(data['feedback'][i] == 'Too many ads') :
            data['feedback'][i] = 8 
            data['feedback_type'][i] = 1   
        else :
            data['feedback'][i] = 9 
            data['feedback_type'][i] = 2
            
            
    print("Done with ........... Feeback ..............\n") 
    print("\n\n Returning data ...............\n")
    
    
    
            
    return data 
    
    
    
    
```

```python
def encode(y) :
    y_encode = list()
    for i in range(len(y)) :
        ilist = list([0,0,0,0,0])
        ilist[y[i]-1] = 1 
        y_encode.append(ilist)
    return np.array(y_encode) 
```

```python
def to_list(encode) :
    val = list()
    for i in range(len(list(encode))) :
        val.append(list(encode[i]).index(max(list(encode[i]))) + 1)
    return val 
```

```python
def get_model() :
    model = Sequential()
    model.add(Conv1D(filters = 32, kernel_size = 3,padding = 'Same', activation ='relu', input_shape = (40,1)))
    model.add(Conv1D(filters = 32, kernel_size = 3,padding = 'Same', activation ='relu', ))
    model.add(MaxPool1D(2))
    model.add(Dropout(0.25))
    model.add(BatchNormalization())


    model.add(Conv1D(filters = 64, kernel_size = 3,padding = 'Same', activation ='relu'))
    model.add(Conv1D(filters = 64, kernel_size = 3,padding = 'Same', activation ='relu'))
    model.add(MaxPool1D(2))
    model.add(Dropout(0.25))
    model.add(BatchNormalization())

    model.add(Conv1D(filters = 128, kernel_size = 3,padding = 'Same', activation ='relu'))
    model.add(Conv1D(filters = 128, kernel_size = 3,padding = 'Same', activation ='relu'))
    model.add(Conv1D(filters = 128, kernel_size = 3,padding = 'Same', activation ='relu'))
    model.add(MaxPool1D(2))
    model.add(Dropout(0.25))
    model.add(BatchNormalization())
    
#     model.add(Conv1D(filters = 256, kernel_size = 3,padding = 'Same', activation ='relu'))
#     model.add(Conv1D(filters = 256, kernel_size = 3,padding = 'Same', activation ='relu'))
#     model.add(Dropout(0.25))
#     model.add(BatchNormalization())
    
    model.add(Flatten())
    model.add(Dense(64))
    model.add(Dropout(0.25))
    model.add(BatchNormalization())
    model.add(Dense(5, activation = "softmax"))
    
    opt = keras.optimizers.Adam()
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    model.summary()
    
    return model 
```

```python
def trainer(train ,y) :
    train = np.array(train)
    y = np.array(y)
    
    model = get_model()
    
    
#     #print(freq)
    strategy_over = {1:4000 , 2:4000}

    
    scaler = StandardScaler()
    kcv = RepeatedStratifiedKFold(n_splits=7, n_repeats=1)
    Folds = kcv.split(train,y)
    counter =  0 
    print(Folds)
    for index_train , index_test in Folds :
        X_train = train[index_train]
        
        y_train = y[index_train]
        X_test = train[index_test]
        y_test = y[index_test]
        
        
        oversample = SMOTE(sampling_strategy=strategy_over)
        X_train, y_train = oversample.fit_resample(X_train, y_train) 
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.fit_transform(X_test)


        X_test = X_test.reshape(X_test.shape[0],X_test.shape[1],1)
        X_train = X_train.reshape(X_train.shape[0],X_train.shape[1],1)
        
        print("Size of training data is ......... ",X_train.shape[0]," and size of test is ",X_test.shape[0],"\n\n")
        

        
        class_weights = list(class_weight.compute_class_weight('balanced',
                                                 np.unique(y_train),
                                                y_train))


        class_weight_dict = dict(enumerate(class_weights))
        
        model.fit(X_train, encode(y_train), validation_data=(X_test, encode(y_test)), 
                  epochs=5, batch_size=32, shuffle=True,
                  class_weight=class_weight_dict)
        

        y_pred = model.predict_classes(X_test)
        y_pred = [i+1 for i in y_pred]
        results_nm = confusion_matrix(y_test,y_pred)
        print(classification_report(y_test,y_pred))
        print(accuracy_score(y_test,y_pred))
        print("Precision Score : ",precision_score(y_test,y_pred, 
                                                   pos_label='positive',
                                                   average='micro'))
        print("Recall Score : ",recall_score(y_test,y_pred, 
                                                   pos_label='positive',
                                                   average='micro'))
        print(f1_score(y_test,y_pred,average='micro'))
                
        
        
        
        
        print("\n",counter," Done ....... and fi_score is ....",f1_score(y_pred,y_test,average="macro"))
        counter += 1
     

    return model 
        
    
    
```
